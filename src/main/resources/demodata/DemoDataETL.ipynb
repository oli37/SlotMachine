### Flights

```python
exp2 = pd.read_csv("20190317_20190321_0000_2359____.exp2", ";", header=None)
exp2.columns = exp2.columns +1
```

```python
exp2.head()
```

```python
selected_date = 190317  #YYMMDD

df = exp2.loc[exp2[9] == selected_date]
df.head()
```

```python
all_flights = exp2.shape[0]
all_flights
```

```python
selected_flights = df.shape[0]
selected_flights
```

```python
(selected_flights/all_flights)*100
```

```python
flights = pd.DataFrame(columns = ['id', 'dep_airport', 'des_airport', 'airline','dep_time', 'des_time'])
flights.loc[:, 'id'] = df.loc[:,8]
flights.loc[:, 'dep_airport'] = df.loc[:,6]
flights.loc[:, 'des_airport'] = df.loc[:,7]
flights.loc[:, 'airline'] = df.loc[:,13]
flights.loc[:, 'dep_time'] = df.loc[:,9]*10000+df.loc[:,10]
#destination time not available, insert dep_time again 
flights.loc[:, 'des_time'] = df.loc[:,9]*10000+df.loc[:,10] 
flights.head()
```

```python
flights.index
```

```python
flights = flights.dropna()
```

```python
flights.shape[0]
```

```python
def dateToTimestamp(date):
    return "TO_TIMESTAMP('"+str(date)+"', 'YYMMDDHH24MI')"
```

```python
flights_sql = []

for i in (flights.index):
    try:
        stmt = "insert into slotmachine.flight values(" \
        + str(flights.loc[i, "id"])                   + ", '"  \
        + str(flights.loc[i, "dep_airport"])          + "', '" \
        + str(flights.loc[i, "des_airport"])          + "', '" \
        + str(flights.loc[i, "airline"])              + "', "  \
        + dateToTimestamp(flights.loc[i, "dep_time"]) + ", "   \
        + dateToTimestamp(flights.loc[i, "des_time"]) + ");"
        
        flights_sql.append(stmt) 
    except:
        pass
```

```python
len(flights_sql)
```

```python
with open('populate_flights.sql', 'w') as f:
    for item in flights_sql:
        f.write("%s\n" % item)
```

### Airports

```python
airports = pd.read_csv("airports.csv", header=None)
airports.head()
```

```python
airports.iloc[:,1] = airports.iloc[:,1].map(lambda x: x.replace("\"", ""))
airports.iloc[:,1] = airports.iloc[:,1].map(lambda x: str(x)).map(lambda x: x.replace("'", ""))
airports.iloc[:,2] = airports.iloc[:,2].map(lambda x: str(x)).map(lambda x: x.replace("'", ""))
airports.iloc[:,3] = airports.iloc[:,3].map(lambda x: str(x)).map(lambda x: x.replace("'", ""))
airports.iloc[:,1] = airports.iloc[:,1].map(lambda x: x.replace(" application.Airport", ""))
airports = airports.iloc[:, [5, 1, 2, 3]]
airports.columns = ['alias','name', 'city', 'country']
airports.head()
```

```python
airports_sql = []
for i in range(airports.shape[0]):
    try:
        stmt = "insert into slotmachine.airport values('" \
        + airports.loc[i, "alias"]    + "', '" \
        + airports.loc[i, "name"]     + "', '" \
        + airports.loc[i, "city"]     + "', '" \
        + airports.loc[i, "country"]  + "');"
        
        airports_sql.append(stmt) 
    except:
        pass
```

### Execute SQL Statements

##### Schema has to be added seperately

```python
def executeSQL(file):
    ok = 0
    nok = 0
    try:
        conn = psycopg2.connect("dbname='postgres' user='postgres' host='localhost' password='postgres'")
        cur = conn.cursor()
    except:
        print ("Unable to connect to DB")
    
    try:
        f = open(file, "r")
    except:
        print("Failed to open File "+file)
        
    for x in f:
        try:
            cur.execute(x)    
            ok+=1
        except:
            nok+=1
        conn.commit()
    print("FILE: "+file+"\tOK: "+str(ok)+"\t\tNOT OK: "+str(nok))
    
```

```python
executeSQL("populate_countries.sql")
```
